```python
docker stop $(docker ps -aq)         #Parar todos os containers ativos
docker rm $(docker ps -aq)           #Remover todos os containers
docker volume prune -f               #Remover volumes persistentes
docker rmi $(docker images -aq) -f   #Remover imagens antigas (Você terá que baixar/buildar tudo novamente depois)
docker builder prune -f              #Limpar cache de build
```
